```python
import math
import time
import os

# ASCII characters from darkest to lightest for depth shading
ascii_shades = " .:-=+*#%@"

def render_sphere(radius, theta, phi):
    width, height = 40, 20
    output = [[" " for _ in range(width)] for _ in range(height)]

    for i in range(0, 360, 10):       # Loop through sphere angles
        for j in range(0, 360, 10):
            x = radius * math.cos(math.radians(i)) * math.sin(math.radians(j))
            y = radius * math.sin(math.radians(i)) * math.sin(math.radians(j))
            z = radius * math.cos(math.radians(j))

            # Rotate around Y-axis
            x_rot = x * math.cos(theta) - z * math.sin(theta)
            z_rot = x * math.sin(theta) + z * math.cos(theta)

            # Rotate around X-axis
            y_rot = y * math.cos(phi) - z_rot * math.sin(phi)
            z_final = y * math.sin(phi) + z_rot * math.cos(phi)

            # Projection to 2D plane
            screen_x = int(width / 2 + x_rot)
            screen_y = int(height / 2 - y_rot)

            # Depth mapping to ASCII characters
            if 0 <= screen_x < width and 0 <= screen_y < height:
                depth_index = int((z_final + radius) / (2 * radius) * (len(ascii_shades) - 1))
                output[screen_y][screen_x] = ascii_shades[depth_index]

    # Print the ASCII sphere
    os.system('cls' if os.name == 'nt' else 'clear')
    for row in output:
        print("".join(row))

def main():
    radius = 8
    theta = 0.0
    phi = 0.0

    try:
        while True:
            render_sphere(radius, theta, phi)
            theta += 0.1  # Slowly rotate around the Y-axis
            phi += 0.05   # Slowly rotate around the X-axis
            time.sleep(0.1)  # Slow down the loop for viewing
    except KeyboardInterrupt:
        print("Exiting program.")

if __name__ == "__main__":
    main()
```
